Your task will be to identify which stimulus was the animal looking at by using the neural activity. For this excersise, we will use the recordings from the primary visual cortex from Adam Kohn's lab - the same dataset as in the previous problem set.

**Excersise 1) Decoding images from spike counts**
<br>
<br>

1) Compute spike counts of each neuron on every trial, split the data into the test/train trials and estimate probabilities $p(r_{n,t}|s_t)$ of spike counts for every neuron and image
<br>
2) Implement the Maximum-Likelihood decoder assuming independence across neurons. What is the performance of the decoder? How much does it exceed the chance level?
<br>
3) Compute the confusion matrix of the Maximum-Likelihood classifier. Which neurons are most frequently confused? Can we say why?
<br>
4) Extra assignement - compute spike counts during different times within a trial and fit the classifier to these different time points - how does it change the performance?

**Excersise 2) Decoding images from time to first spike**
<br>
<br>

Repeat the same excersise as above, but replace spike counts with times to the first spike for each neuron on each trial. How do the two classifiers compare? What does it tell us about representation of information in the primary visual cortex?
