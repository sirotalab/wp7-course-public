**Excersise 1) Clustering neural responses**
<br>
<br>

1) Use the provided data matrix of > 30 000 RGC responses to a visual "chirp" stimulus (file chirp_response.npy). Load it, visualize some of the neurons. Do you see any signatures of a clustering already?
<br>
2) Using the sklearn package implement the Gaussian mixture model.  Split the data into a test and train set. Fit (using the diagonal covariance) different numbers of clusters and compute a goodness-of-fit measure (such as the log likelihood) on the test set. Is any number of clusters preferred? Remember to visualize cluster centroids and activity of neurons within each cluster.
<br>
3) Change the variant of the GMM - fit it using the diagonal covariance, shared covariance and full covariance matrices. Does it affect the results? Plot covariance matrices for selected clusters.
<br>
4) Discuss - what conclusions can we draw about functional types of RGCs based on this analysis?
